```python
from pydub import AudioSegment
import numpy as np

# Ses dosyasını yükleyin
audio = AudioSegment.from_wav("/kaggle/input/sample-wav/sample-3s.wav")

# Ses dosyasındaki örnekleri numpy dizisine dönüştürün
samples = np.array(audio.get_array_of_samples())

# Gizli mesajı binary formata çevirin
message = "Merhaba, bu bir test mesajidir!"
binary_message = ''.join(format(ord(c), '08b') for c in message)

# LSB algoritması ile mesajı gizleme fonksiyonu
def lsb_insert(audio_samples, message_binary):
    message_index = 0
    for i in range(len(audio_samples)):
        if message_index < len(message_binary):
            # Ses örneğinin son bitini gizli mesajın bitine yerleştiriyoruz
            audio_samples[i] = (audio_samples[i] & ~1) | int(message_binary[message_index])
            message_index += 1
    return audio_samples

# LSB ile mesajı gizleyin
modified_samples = lsb_insert(samples, binary_message)

# Gizlenmiş ses dosyasını kaydedin
modified_audio = audio._spawn(modified_samples.tobytes())

# Kaydetme işlemi
output_path = "/kaggle/working/output_audio.wav"
modified_audio.export(output_path, format="wav")

# Gizlenmiş ses dosyasının kaydedildiğini kontrol edelim
print(f"Gizlenmiş ses dosyası '{output_path}' olarak kaydedildi.")
```

```python
import numpy as np
from pydub import AudioSegment

# Gizlenmiş ses dosyasını yükleyin
audio = AudioSegment.from_wav("/kaggle/working/output_audio.wav")

# Ses dosyasındaki örnekleri numpy dizisine dönüştürün
samples = np.array(audio.get_array_of_samples())

# LSB algoritması ile gizli mesajı çıkarma fonksiyonu
def lsb_extract(audio_samples, message_length):
    binary_message = ""
    for i in range(message_length):
        # Son biti alıyoruz (gizlenmiş mesajın bitini)
        binary_message += str(audio_samples[i] & 1)
    return binary_message

# Gizlenmiş mesajın uzunluğu
message_length = len("Merhaba, bu bir test mesajidir!") * 8  # Her karakterin 8 bit olduğunu varsayarak

# Gizli mesajı çıkarın
binary_message = lsb_extract(samples, message_length)

# Binary mesajı metne dönüştürün
message = ''.join(chr(int(binary_message[i:i+8], 2)) for i in range(0, len(binary_message), 8))

# Çıkarılan mesajı yazdırın
print("Gizlenmiş mesaj:", message)
```

```python
original_message = "Merhaba, bu bir test mesajidir!"
if message == original_message:
    print("Gizlenmiş mesaj başarıyla çıkarıldı ve doğru!")
else:
    print("Bir hata var, gizlenen mesaj doğru çıkarılmadı.")
```
